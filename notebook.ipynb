```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

categorical_pipeline = Pipeline(
            [
                ("simple_imputer", SimpleImputer(missing_values=np.nan, strategy="constant", fill_value=None, add_indicator=True)),
                ("one_hot_encoder", OneHotEncoder(handle_unknown="ignore")),
            ]
        )

numerical_pipeline = Pipeline(
            [
                ("simple_imputer", SimpleImputer(missing_values=np.nan, strategy="constant", fill_value=None, add_indicator=True)),
                ("scaler", MinMaxScaler()),
            ]
        )
```

```python
numerical_pipeline[1].data_max_ - numerical_pipeline[1].data_min_

class MinMaxScalerReverse:
    def __init__(self, minval, maxval):
        self.minval = minval
        self.maxval = maxval
        
    def inverse_transform(self, x):
        return x * (self.maxval - self.minval) + self.minval
    
```

```python
data = pd.DataFrame({
    'x': [1, 2, 3, 4, 5],
    'y': [1, 2, 3, 4, 5]
})

numerical_pipeline.fit(data)
scaled_data = numerical_pipeline.transform(data)

numerical_pipeline[1].data_max_

reverse_transformer = MinMaxScalerReverse(minval = 1, maxval = 5)

data['x'].max()

# numerical_pipeline.inverse_transform(scaled_data)
```
